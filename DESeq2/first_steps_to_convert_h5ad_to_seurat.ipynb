```python
#h5ad files is provided on Figshare
ndata = sc.read('/path/25_11_22_Xenium_Dataset1_290_IntReps1and2_Annotated.h5ad')
```

```python
ndata.X = ndata.layers['raw_counts']
```

```python
ndata
```

```python
with open ('matrix_files/barcodes.tsv' , 'w') as f:
    for item in ndata.obs_names:
        f.write(item + '\n')
```

```python
with open('matrix_files/features.tsv' , 'w') as f:
    for item in ['\t' .join([x,x, 'Gene Expression']) for x in ndata.var_names]:
        f.write(item + '\n')
```

```python
io.mmwrite('matrix_files/matrix', ndata.X.T)
```
